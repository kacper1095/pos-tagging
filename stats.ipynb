```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from collections import Counter
```

```python
data = np.load(Path('data') / 'raw' / 'competition_dataset.npy')
labels = np.load(Path('data') / 'raw' / 'competition_dataset_labels.npy')
test = np.load(Path('data') / 'raw' / 'competition_test.npy')
```

```python
max([len(sample) for sample in data])
```

```python
flat_labels = []
flat_words = []
flat_test = []
for l in labels:
    flat_labels.extend(l)
for s in data:
    flat_words.extend(s)
for s in test:
    flat_test.extend(s)
plt.hist(flat_labels, np.unique(flat_labels))
```

```python
counts = Counter(list(zip(flat_words, flat_labels)))
flat_frame = pd.DataFrame(data={
    "word": flat_words,
    "labels": flat_labels,
})
```

```python
len(np.unique(flat_words))
```

```python
counts.most_common(50)
```

```python
interp = flat_frame[flat_frame['labels'] == 1]
unique_interp = np.unique(interp['word'])
unique_interp
```

```python
test_interp = []
for word in flat_test:
    for interp in unique_interp:
        if interp in word:
            test_interp.append(word)
list(np.unique(test_interp))
```

```python
100 - len(flat_test) / len(set(flat_test).difference(set(flat_words)))
```
